# Data Cleaning in Pandas 

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\irvin\OneDrive\Desktop\ARAL SERIES\DATA ANALYTICS\PANDAS PYTHON\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#strip can work on a 1 condition
df["Last_Name"] = df["Last_Name"].str.strip("._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex =True)
df
```

```python
#Change data type to str before applying lambda condition
#other syntax: df['Phone_Number'] = df['Phone_Number'].astype(str)
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
df
```

```python
# Replace values in both 'Paying Customer' and 'Do_Not_Contact' columns in a single line code
df[['Paying Customer', 'Do_Not_Contact']] = df[['Paying Customer', 'Do_Not_Contact']].replace({'N': 'No', 'Y': 'Yes'})
df
```

```python
# Replace both 'N/a' and 'NaN' with an empty string in a single step
df = df.replace(['N/a', 'NaN'], '')

#To fill blanks that does not have value basically replacing it with something
df=df.fillna('')
df
```

```python
# Filter out rows where "Do_Not_Contact" is 'Yes'
df = df[df["Do_Not_Contact"] != 'Yes']

# Display the updated DataFrame
df

#Other way
#for x in df.index:
#    if df.loc[x, "Do_Not_Contact"] == 'Y':
#        df.drop(x, inplace=True)
```

```python
# Filter out rows where "Do_Not_Contact" is 'Yes'
df = df[df["Phone_Number"] != '']

# Display the updated DataFrame
df

#Other ways to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)

# for x in df.index:
#     if df.loc[x, "Phone_Number"] == '':
#         df.drop(x, inplace=True)
```

```python
#To reset index into new index
df = df.reset_index(drop=True)
df
```
