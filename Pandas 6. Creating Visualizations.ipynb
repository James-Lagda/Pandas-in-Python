# Creating Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\irvin\OneDrive\Desktop\ARAL SERIES\DATA ANALYTICS\PANDAS PYTHON\Ice Cream Ratings.csv")
df = df.set_index('Date')
df               
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s= 150, title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s= 150, c = 'red', title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.hist(title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.hist(bins = 20, title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.boxplot()
```

```python
df.plot.area(title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.area(figsize = (10,5), title = 'Ice Cream Ratings', xlabel = 'Daily Rate', ylabel = 'Scores')
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,10), title = 'Ice Cream Ratings')
```
